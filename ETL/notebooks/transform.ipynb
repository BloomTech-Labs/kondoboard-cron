```python
import pandas as pd
from bs4 import BeautifulSoup
import arrow
```

```python
df = pd.read_csv('./data/adzuna_extract.csv')
```

```python
df.head()
```

```python
def keyword(text):
    """
    Formats keywords by 
    -lowercasing all letters 
    -replacing ' ' with '_'
    -replacing '-' with '_'
    """
    text = text.lower()
    text = text.replace(' ', '_')
    text = text.replace('-', '_')
    return text
```

```python
def remove_html(text):
    """
    Removes HTML
    """
    return BeautifulSoup(text).get_text()
```

```python
def format_date(text):
    """
    Formats Date to 'YYYY-MM-DD'
    """
    time = arrow.get(text)
    time = time.format('YYYY-MM-DD')
    return time
```

```python
def adzuna_id(text):
    """
    Formats ADZUNA API IDs to A + original ID
    """
    return "A" + str(text)
```

```python
def apply(column, function):
    """
    Applies function to column
    """
    df[column] = df[column].apply(function)
```

```python
# 
columns_to_remove_html=["title", "title_keyword"]

for col in columns_to_remove_html:
    df[col] = df[col].apply(remove_html)
```

```python
# InsertedDate 

df['inserted_date'] = arrow.utcnow().format("YYYY-MM-DD")
```

```python
# APPLY ADZUNA ID

apply('id', adzuna_id)
```

```python
# APPLY REMOVE_HTML FUNCTION

apply('title', remove_html)
apply('title_keyword', remove_html)
```

```python
# APPLY KEYWORD FUNCTION

apply('title_keyword', keyword)
apply('tags', keyword)  
```
